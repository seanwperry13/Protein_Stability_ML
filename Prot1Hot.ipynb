```python
import pandas as pd
import os

file_dir1='/data/sd2/versioned-datasets/data/protein-design/metadata/'
all_files1=os.listdir(file_dir1)

df0=pd.DataFrame()

for j in range(len(all_files1)):
    file_name1 = all_files1[j]
    unfiltered_csv1 = pd.read_csv(file_dir1+file_name1, comment='#')
    df1=pd.DataFrame(unfiltered_csv1)
    if 'sequence' in df1.columns and 'name' in df1.columns and 'topology' in df1.columns and 'dataset' in df1.columns:
        df0=df0.append(df1[['dataset','name','sequence','topology']].dropna(), ignore_index = True)
        

file_dir2='/data/sd2/versioned-datasets/data/protein-design/experimental_stability_scores/'
all_files2=os.listdir(file_dir2)

df2=pd.DataFrame()

for j in range(len(all_files2)):
    file_name2 = all_files2[j]
    unfiltered_csv2 = pd.read_csv(file_dir2+file_name2, comment='#')
    df3=pd.DataFrame(unfiltered_csv2)
    if 'stabilityscore_cnn_calibrated' in df3.columns and 'name' in df3.columns and 'dataset' in df3.columns:
        df2=df2.append(df3[['dataset','name','stabilityscore_cnn_calibrated',]].dropna(), ignore_index = True)
 
df=df0.merge(df2,'inner')
df
```

```python
import numpy as np
len(df['sequence'].values)
length_checker = np.vectorize(len)
seq_length = length_checker(df['sequence'].values)
N=max(seq_length)
M=len(df)

L=np.full((M,N),'0')
np.shape(L)

for i in range(len(df)):
    for j in range(len(df['sequence'].values[i])):
        L[i][j]=list(df['sequence'].values[i])[j]
```

```python
for i in range(N):
    pos_lbl='p'+str(i+1)
    df[pos_lbl]=L[:,i]
df
```

```python
from sklearn.preprocessing import OneHotEncoder

data=df.drop(columns=['dataset','name','sequence','topology','stabilityscore_cnn_calibrated'])

enc = OneHotEncoder(categories='auto') 
enc.fit(data)
D = enc.transform(data).toarray()
X = pd.DataFrame(D)
y = df['stabilityscore_cnn_calibrated']
```

```python
from sklearn.ensemble import RandomForestRegressor

regr = RandomForestRegressor(n_estimators=100,n_jobs=10)
regr.fit(X, y)
```

```python
abs_dif = abs(y.values-regr.predict(X))
np.mean(abs_dif)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
regr0 = RandomForestRegressor(n_estimators=100,n_jobs=10)
regr0.fit(X_train, y_train)
```

```python
regr1 = RandomForestRegressor(n_estimators=100,n_jobs=10)
regr1.fit(X_test, y_test)
```

```python
abs_dif0 = abs(y_train.values-regr.predict(X_train))
print(np.mean(abs_dif0))

abs_dif1 = abs(y_test.values-regr.predict(X_test))
print(np.mean(abs_dif1))
```

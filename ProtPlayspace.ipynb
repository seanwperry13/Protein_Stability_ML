```python
import numpy as np
import gudhi as gd
ac = gd.AlphaComplex(points=df.values)

st = ac.create_simplex_tree()
result_str = 'Alpha complex is of dimension ' + repr(st.dimension()) + ' - ' + \
    repr(st.num_simplices()) + ' simplices - ' + \
    repr(st.num_vertices()) + ' vertices.'
print(result_str)

fmt = '%s -> %.2f'
for filtered_value in st.get_filtration():
    print(fmt % tuple(filtered_value))

points = np.array([ac.get_point(i) for i in range(st.num_vertices())])
# We want to plot the alpha-complex with alpha=0.005 by default.
# We are only going to plot the triangles
triangles = np.array([s[0] for s in st.get_skeleton(4) if len(s[0])==4 and s[1] <= 3])

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

fig = plt.figure()
ax = fig.gca(projection='3d')
l = ax.plot_trisurf(points[:,0], points[:,1], points[:,2], triangles=triangles)
ax.set_xlim(-6, 16)
ax.set_ylim(0, 25)
ax.set_zlim(-15, 15)

plt.show()
```

```python
# define a function to identify single site mutations from amino acid sequences
def diffby1(str1, newstr):
    d = 0
    i = 0
    swapid = ''
    for i in range(min(len(str1), len(newstr))):
        if str1[i] != newstr[i]:
            d += 1
            swapid = '%s%d%s' % (str1[i], i+1, newstr[i])
        if d > 1:
            return (False, '')
    return (d == 1, swapid)
# load the experimental stability scores dataframe
sequences_df = pd.read_csv('/data/sd2/protein-design/ssm_stabilities/ssm_rescues.v3.experimental_stability_scores.csv', comment='#')
# separate out ladder designs
ladder_df = sequences_df[sequences_df['name'].apply(lambda row: 'ladder' in row)].copy()
sequences_df = sequences_df[sequences_df['name'].apply(lambda row: 'ladder' not in row)].copy()
sequences_df.reset_index(drop=True, inplace=True)
# load the original metadata for the 19 Rocklin designs in this experiment
rd1ssm_df = pd.read_csv('/data/sd2/protein-design/ssm_stabilities/rd1_unexpectedly_unstable_designs.csv', comment='#', usecols=['name', 'sequence'])
# loop over each design and find the corresponding ssm designs in the experiment
sequences_df['new_name'] = ''
for i in range(len(rd1ssm_df)):
    name = rd1ssm_df.loc[i, 'name']
    oseq = rd1ssm_df.loc[i, 'sequence']
    for j in range(len(sequences_df)):
        if len(sequences_df.loc[j, 'new_name']) == 0:
            nseq = sequences_df.loc[j, 'protein_sequence']
            swapflag, swapid = diffby1(oseq, nseq)
            if oseq in nseq:
                # exact match to original design, use original name
                sequences_df.loc[j, 'new_name'] = name
            elif swapflag:
                # found a ssm, encode ssm in original name
                sequences_df.loc[j, 'new_name'] = name + '_' + swapid
    print('Finished %s' %name)
    
    
```
